```python
import numpy as np

t = np.array([660, 630, 660, 720])

t_ges = np.sum(t)
T = t_ges / t.size
print(T)

dt_syst = 15 # s
dt_streu = np.sqrt(5 * (1 / 5 * dt_syst)**2)
print(dt_streu)
```

```python
L = 4.15 # m
m_1 = 1.5 # kg

b = 49.4 # mm
b /= 1000 # mm -> m

d = 50 # mm
d /= 1000 # mm -> m

A = 15.7 # cm
A /= 100 # cm -> m

delta_A = 0.7 # cm
delta_A /= 100 # cm -> m



delta_G_tmp = ((np.pi**2*d*b**2)/(m_1*T**2*L)*delta_A)**2
delta_G_tmp += ((2 * np.pi**2 * b**2 * A * d)/(m_1 * T**3 * L) * dt_streu)**2
delta_G = np.sqrt(delta_G_tmp)

print("Delta_G:", delta_G)


G = (np.pi**2 * b**2 * A * d)/(m_1 * T**2 * L)
print("Calculated G", G)
print("Delta_G in %:", delta_G / G * 100, "%")

real_G = 6.67384E-11
print("Delta to Real G in %:", (G - real_G) / real_G * 100, "%")

real_G = 6.67259E-11
print("Delta to Real G in %:", (G - real_G) / real_G * 100, "%")
```

```python

delta = (b**3)/((b**2 + 4*d**2)**(3/2))
print("Delta:", delta * 100 , "%")

G_mod = G * (1/(1-delta))

print("Mod_G", G_mod)

delta_G_mod = delta_G * (1/(1-delta))

print("Delta_G_mod", delta_G_mod)

print("Delta_G_mod delta to Delta_G", delta_G_mod / delta_G * 100, "%")
print("Delta_Mod_G", delta_G_mod / G_mod * 100, "%")

print("Delta to Real G in %:", (G_mod - real_G) / real_G * 100, "%")
```
