```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 100)  # Sample data.
np.seterr(divide='ignore', invalid='ignore')


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, 1/np.sqrt(x+1), label='original')
plt.plot(x, 1-x/2+(3*x**2)/8, label='taylor')
plt.xlabel('x')
plt.ylabel('y')
plt.title("$f_1(x)$")
plt.legend()
plt.savefig("output/Aufgabe 1 - f_1(x).png")

plt.show()
```

```python
x = np.linspace(0.5, 1, 100)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.plot(x, np.sin(np.log(x**2)), label='original')  # etc.
plt.plot(x, 2*(x-1)-(x-1)**2-2/3*(x-1)**3, label='taylor')
plt.xlabel('x')
plt.ylabel('y')
plt.title("$f_3(x)$")
plt.legend( loc = 'lower right')
plt.savefig("output/Aufgabe 1 - f_3(x).png")

plt.show()
```

```python
x = np.linspace(1, 3, 100)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.plot(x, x**4-3*x**3+7*x-1, label='original')  # etc.
plt.plot(x, 5 + 3*(x-2) + 6*(x-2)**2 + 5*(x-2)**2 + (x-2)**4, label='taylor')
plt.xlabel('x')
plt.ylabel('y')
plt.title("$f_4(x)$")
plt.legend()
plt.savefig("output/Aufgabe 1 - f_4(x).png")

plt.show()
```

```python
x = np.linspace(1, 4, 100)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.plot(x, 1/x, label='original')  # etc.
plt.plot(x, 1/2 - 1/4*(x-2) + 1/2*2/8*(x-2)**2 - 1/6*6/(2**4)*(x-2)**3 + 1/24*24/(2**5)*(x-2)**4, label='taylor')
plt.xlabel('x')
plt.ylabel('y')
plt.title("$f(x)$")
plt.legend()
plt.savefig("output/Aufgabe 2 - f(x).png")

plt.show()
```

```python
x = np.linspace(1, 4, 100)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.plot(x, np.absolute(1/x - (1/2 - 1/4*(x-2) + 1/2*2/8*(x-2)**2 - 1/6*6/(2**4)*(x-2)**3 + 1/24*24/(2**5)*(x-2)**4)), label='$d_N(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.title("$d_N(x)$")
plt.legend()
plt.savefig("output/Aufgabe 2 - d_N(x).png")

plt.show()
```
